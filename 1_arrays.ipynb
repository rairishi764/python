```python

#Arrays are stored contiguously in memory
# 2D arrays also are stored linearly row wise
# array is not built in data model in python, list is
# array is std python module and more efficient than list
from array import *
a1 = array('i',[1,2,3,4,5]) # O(n)
print(a1)
            
```

```python
#insert shifts the array elements and inserts in middle
a1.insert(6,9) # O(n) complexity as moving of elements happens 
print(a1)
a1.insert(0,4)
a1.remove(3) # removes 1st occurance fo element in array, O(n) as has to shift all latter elements by 1 place
print(a1)
```

```python
# traversal
for i in a1:  # O(n) 
    print (i, end= ' ')
```

```python
def linear_search(array, value):
    print (array)
    for i in array:
        if i==value:
            return array.index(value) #index of given value sent
    
    return -1

print(linear_search(a1, 5))
```
